9. Determine el promedio de edad de quienes cursan una carrera régimen diurno versus vespertino.

```python
df["fn"] = df["fec_nac_alu"].astype(str).apply(lambda x: x[4:6] + "-14-" + x[0:4])
df["fn"] = pd.to_datetime(df["fn"])
```

```python
import numpy as np
```

```python
fecha_original = pd.to_datetime("2023-03-01")
df["edad"] = df["fn"].apply(lambda x: np.floor((fecha_original - x).days / 365))
```

```python
df.groupby("jornada").agg({"edad": "mean"})
```
