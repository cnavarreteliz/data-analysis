5. Determine las 10 carreras de mayor costo en arancel en el gran Concepción. Considere un valor UF estimado de $36.000.


```python
df["formato_valores"].unique()
```

```python
df["valor2"] = df.apply(lambda x: x["valor_arancel"] * 36000 if x["formato_valores"] == "Monto en UF" else x["valor_arancel"], axis=1)
```

```python
df["comuna_sede"].unique()
```

```python
comunas = ["CONCEPCION", "HUALPEN", "SAN PEDRO DE LA PAZ", "TALCAHUANO", "CHIGUAYANTE"]
```

```python
df_5 = df[df["comuna_sede"].isin(comunas)]
df_5.head()
```

```python
df_5.groupby("nomb_carrera").agg({"valor2": "mean"}).sort_values("valor2", ascending=False).head(10)
```
